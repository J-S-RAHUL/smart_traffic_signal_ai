```
import numpy as np
import random


class TrafficSignalEnv:
    def __init__(self):
        self.state = np.zeros(4, dtype=np.int32)
        self.emergency_vehicles = [0, 0, 0, 0]

    def reset(self):
        self.state = np.random.randint(5, 20, size=4)
        self.emergency_vehicles = [
            random.choice([0, 0, 0, 1]) for _ in range(4)
        ]
        return self.state

    def step(self, action):
        reward = 0
        road_names = ["North", "South", "East", "West"]

        max_traffic_road = np.argmax(self.state)

        emergency_waiting = any(self.emergency_vehicles)
        emergency_road = (
            self.emergency_vehicles.index(1)
            if emergency_waiting else -1
        )

        if emergency_waiting and action != emergency_road:
            reward -= 10

        if action == max_traffic_road:
            reward += 10
        else:
            reward -= 3

        waiting_penalty = np.sum(self.state) // 10
        reward -= waiting_penalty

        reduction = int(self.state[action] * 0.8)
        self.state[action] = max(
            0, self.state[action] - reduction
        )

        arrivals = np.random.randint(1, 6, size=4)
        self.state = np.clip(self.state + arrivals, 0, 50)

        return self.state, reward, road_names[action]


env = TrafficSignalEnv()
obs = env.reset()

print("🚦 Smart Traffic Signal AI Simulation")
print("Initial Traffic:", obs)
print("-" * 30)

for i in range(5):
    action = np.argmax(obs)
    obs, reward, road = env.step(action)

    print(f"Step {i+1}")
    print(f"Green Signal: {road}")
    print(f"Reward: {reward}")
    print(f"Traffic: {obs}")
    print("-" * 30)
```
